```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('dataAug')
```

```python
def pre_prediction(imag):
    img = cv2.resize(imag, (64,64))
    img = img.reshape((1,64,64,1)) / 255.0
    p = model.predict(img)[0]
    return p.argmax()
```

```python
cnt = 0
prd = []
for i in range(2062):
    p = pre_prediction(X[i])
    prd.append(p)
    if p == y[i]:
        cnt += 1
```
